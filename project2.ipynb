```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
class LLMState(TypedDict):
    que: str
    ans: str
```

```python
def llm_agent(state: LLMState) -> LLMState:
    question = state['que']

    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
    prompt = f'''
    You have to answer a question present at the end.

    Instructions:
    1. Give concise answer.
    3. Answer in a friendly and encouraging tone.

    Question: {question}
    '''

    answer = llm.invoke(prompt).content
    
    state['ans'] = answer
      
    return state
```

```python
graph = StateGraph(LLMState)

graph.add_node('llm_agent', llm_agent)

graph.add_edge(START, 'llm_agent')
graph.add_edge('llm_agent', END)

workflow = graph.compile() 
```

```python
input_state = {'que': 'Why does earth and other planets follow an elliptical path around the sun?'}
output_state = workflow.invoke(input_state)

print(output_state['ans'])
```
